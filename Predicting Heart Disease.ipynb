## Predictive Modelling

### 1. Logistic Regression

```python
import os
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
categorical_columns = ['Gender', 'Family History', 'Diabetes', 'Obesity', 'Exercise Induced Angina']
heart = pd.get_dummies(df, columns=categorical_columns, drop_first=True)
```

```python
heart.head()
```

```python
le = LabelEncoder()
heart['Smoking'] = le.fit_transform(heart['Smoking'])
heart['Alcohol Intake'] = le.fit_transform(heart['Alcohol Intake'])
heart['Chest Pain Type'] = le.fit_transform(heart['Chest Pain Type'])
```

```python
heart.head()
```

```python
# Define features (X) and target (y)
X = heart.drop('Heart Disease', axis=1)
y = heart['Heart Disease']
```

```python
# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Scale numerical features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Train Logistic Regression model
model = LogisticRegression(random_state=42)
model.fit(X_train, y_train)
```

```python
# Predict and evaluate
y_pred = model.predict(X_test)
y_prob = model.predict_proba(X_test)[:, 1]
```

```python
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
print("Accuracy:", accuracy)
print("Confusion Matrix:\n", conf_matrix)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
import statsmodels.api as sm

# 1. McFadden's R²
# Add a constant to the feature set for statsmodels
X_test_with_const = sm.add_constant(X_test)
X_train_with_const = sm.add_constant(X_train)

# Fit logistic regression model using statsmodels
sm_model = sm.Logit(y_train, X_train_with_const).fit()

# Log-likelihood of the fitted model
log_likelihood_model = sm_model.llf

# Log-likelihood of the null model (intercept-only)
log_likelihood_null = sm_model.llnull

# McFadden's R²
pseudo_r2 = 1 - (log_likelihood_model / log_likelihood_null)
print(f"McFadden's R²: {pseudo_r2:.4f}")

# 2. Mean Absolute Error (MAE)
# Use predicted probabilities
mae = mean_absolute_error(y_test, y_prob)
print(f"Mean Absolute Error (MAE): {mae:.4f}")

# 3. Root Mean Squared Error (RMSE)
rmse = np.sqrt(mean_squared_error(y_test, y_prob))
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
```

```python
# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")

# Classification report
print("Classification Report:")
print(classification_report(y_test, y_pred))

# Confusion matrix
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

### 2. KNN Model

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn.impute import SimpleImputer
```

```python
df['Gender'] = df['Gender'].map({'Male': 0, 'Female': 1})
df['Smoking'] = df['Smoking'].map({'Never': 0, 'Former': 1, 'Current': 2})
df['Alcohol Intake'] = df['Alcohol Intake'].map({'None': 0, 'Moderate': 1, 'Heavy': 2})
df['Family History'] = df['Family History'].map({'No': 0, 'Yes': 1})
df['Diabetes'] = df['Diabetes'].map({'No': 0, 'Yes': 1})
df['Obesity'] = df['Obesity'].map({'No': 0, 'Yes': 1})
df['Exercise Induced Angina'] = df['Exercise Induced Angina'].map({'No': 0, 'Yes': 1})
df['Chest Pain Type'] = df['Chest Pain Type'].map({'Asymptomatic': 0, 'Typical Angina': 1, 'Atypical Angina': 2, 'Non-anginal Pain': 3})
```

```python
X = df.drop('Heart Disease', axis=1)
y = df['Heart Disease']
```

```python
imputer = SimpleImputer(strategy='median')  
X = imputer.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
knn = KNeighborsClassifier(n_neighbors=3,
                          #metric= 'eucledian'
                          )
```

```python
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
```

```python
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
print("\nAccuracy Score:", accuracy_score(y_test, y_pred))
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
import pandas as pd

def calculate_vif(X):
    vif_data = pd.DataFrame()
    vif_data["Feature"] = X.columns
    vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    return vif_data

scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)

vif_data = calculate_vif(X_scaled)

print(vif_data)
```
